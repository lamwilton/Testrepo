```python
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X = np.arange(100)
X = np.vstack((X, np.arange(200, 300))).T
y = np.arange(500, 600).reshape(-1, 1)
# x_train, y_train, x_test, y_test = train_test_split(X, y, test_size=0.2)
TEST_SIZE = 0.2
time_train = TimeseriesGenerator(X, y, length=5, end_index=round(X.shape[0] * TEST_SIZE))
time_test = TimeseriesGenerator(X, y, length=5, start_index=round(X.shape[0] * TEST_SIZE))
xg, yg = time_train[0]
xt, yt = time_test[0]

print(xg)
print(yg)
print(xt)
print(yt)
```
